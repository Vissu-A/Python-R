```python
import pandas as pd
import datetime
import os

df = pd.read_csv(r'C:\Users\vadapala\Desktop\IBM_SERVER_TIMESERIES_CODE\iHYDIBM_LP.csv')

df.drop(['Unnamed: 11'], axis =1, inplace = True)

print(df.head())

print(list(df))
```

```python
grouped = df.groupby('Node_Name')

for x in grouped:
    
    name = x[0].strip()
    print(name)
    print()
    dframe = x[1]
    print(dframe.head())
    print('============================================================================')
    
    dir_name = 'Grourby_dataframe_file'
    if not os.path.exists(dir_name):
        os.makedirs(dir_name)
    filename = name+'.csv'
    
    path = os.path.join(dir_name, filename)
    
    dframe.to_csv(path, index=False)
```

```python
files = []
for f in os.listdir(dir_name):
    if os.path.isfile(os.path.join(dir_name,f)):
        files.append(f)

print(files)
print(len(files))
```
